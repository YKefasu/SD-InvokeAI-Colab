```python
#@markdown #Step 3Ô∏è‚É£: Launch InvokeAI WebUI [`Colab UPDATE: 18 Oktober 2023 20:00 PM (GMT+7)`]
#@markdown ______
import os
import threading
import requests
import logging
import subprocess
from pyngrok import ngrok
from IPython.display import clear_output, display, HTML

# Function AntiDisconnect in the background
clear_output()
audio_url = "https://raw.githubusercontent.com/KoboldAI/KoboldAI-Client/main/colab/silence.m4a"

def play_audio(url):
    display(HTML(f'<audio src="{url}" controls autoplay style="display:none"></audio>'))

audio_thread = threading.Thread(target=play_audio, args=(audio_url,))
audio_thread.start()

#@markdown Copy AuthToken dari https://dashboard.ngrok.com/get-started/your-authtoken ke üëá  ( WAJIB DI ISI üëá)
ngrok_token = ""  # @param {'type': 'string'}

# Disable logging for the pyngrok module to suppress the warnings
logging.getLogger("pyngrok").setLevel(logging.ERROR)
os.system("mv /root/.config/ngrok/ngrok.yml /root/.ngrok2/ngrok.yml")
os.environ["NGROK_CONFIG_PATH"] = "/root/.ngrok2/ngrok.yml"
ngrok_ready_event = threading.Event()

# Function to install ngrok via TGZ file
def install_cloudflare():
    !wget https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O cloudflared
    !chmod a+x cloudflared

install_cloudflare()
f = open("stdout", "w")
p = subprocess.Popen(['./cloudflared', '--url', 'http://127.0.0.1:9090'], bufsize=0, stdout=f, stderr=subprocess.STDOUT)

# Function to start ngrok and print the ngrok URL when the local URL is accessible
def start_ngrok():
    ngrok.set_auth_token(ngrok_token)
    public_url = ngrok.connect(9090, "http").public_url
    public_url_https = public_url.replace("http://", "https://", 1)  # Use count=1 to replace only the first occurrence

    # Print the ngrok URL once the local URL is accessible
    url_printed = False
    while True:
        try:
            response = requests.get("http://127.0.0.1:9090")
            if response.status_code == 200 and not url_printed:
                print("+++ Subscribe to: https://youtube.com/@yosiakefas")
                print("Ngrok URL:", public_url_https)
                print("+++")
                !grep -F trycloudflare stdout

                # Start Cloudflare Tunnel here
                os.system(f'./cloudflared tunnel --url {public_url}')

                url_printed = True  # Set the flag to indicate that the URL has been printed
                break
        except requests.ConnectionError:
            pass

    ngrok_ready_event.set()  # Signal that ngrok tunnel is ready

# Install ngrok via TGZ file in a separate thread
install_thread = threading.Thread(target=install_ngrok)
install_thread.start()
ngrok_ready_event.wait()

# Create a Timer to allow the main thread to continue after 5 seconds
timer = threading.Timer(1, lambda: None)  # A dummy timer function to unblock the main thread after 5 seconds
start_thread = threading.Thread(target=start_ngrok)
start_thread.start()

# Start the timer to unblock the main thread after 5 seconds
timer.start()
timer.join()

# Continue with the rest of your code
clear_output()
%env LD_PRELOAD=/usr/lib/x86_64-linux-gnu/libjemalloc.so.2
!python /content/InvokeAI/scripts/invokeai-web.py --root /content/sd-db

# Kill ngrok when done
ngrok.kill()
```

```python
#@markdown #Copy Output to Google Drive >  InvokeAI-Output

import os
import shutil

# Define source and destination paths
source_folder = "/content/sd-db/outputs/images"
destination_folder = "/content/drive/MyDrive/InvokeAI-Output/"

# Check if the source folder exists
if not os.path.exists(source_folder):
    print("Source folder does not exist.")
else:
    # Check if the destination folder exists, create if it doesn't
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate through the files and subfolders in the source folder
    for item in os.listdir(source_folder):
        source_item_path = os.path.join(source_folder, item)
        destination_item_path = os.path.join(destination_folder, item)

        # Check if the item is a file
        if os.path.isfile(source_item_path):
            # Check if the item already exists in the destination folder
            if os.path.exists(destination_item_path):
                print(f"Overwriting {item}")

            # Copy the file to the destination folder, overwriting if necessary
            shutil.copy2(source_item_path, destination_item_path)
        elif os.path.isdir(source_item_path):
            # Copy directories recursively to the destination folder, merging if necessary
            shutil.copytree(source_item_path, destination_item_path, dirs_exist_ok=True)

    print("Folder copy completed.")
```
